# 1. Описание данных

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# выгрузим таблицу и посмотрим структуру данных
df = pd.read_csv('/content/telecom_users.csv')
df.head()
```

```python
# посмотрим на пустые значения
df.replace(['', ' ', '  '], np.nan).isnull().sum() 
```

```python
# отлично, пропусков почти нет, только 10 значений в столбце с суммой всех платежей
# здесь мы видим что пустые значения есть только для тех клиентов, кто сотрудничает первый месяц
# соответственно у них эта сумма просто равна нулю
df[df['TotalCharges'].isin([' ', '  ', np.nan])==True]
```

```python
# посмотрим основную статистику числовых признаков
df[['TotalCharges', 'MonthlyCharges', 'tenure']].replace(' ', 0).astype('float').describe()
```

Мы видим, что в датасете среднее время работы клиента с провайдером - 2 года и 8 месяцев.При этом разброс значений такой, что в стандартное отклонение входят клиенты со стажем от 8 мес. и до 4 лет и 8 мес. Похожая ситуация с размером оплаты. Для общей суммы платежей стандартное отклонение, практически, равно среднему, а для ежемесячного списания половине от среднего. Таким образом в обоих случаях среднее минус и плюс станд. отклонение находится ниже первого квартиля всех значений и выше третьего.

```python
# чтобы не перезапускать первый параграф скопируем дф
data = df.copy()
# переопределяем категории в цифровые значения, простая разметка
replace_dict = {'Yes': 1, 'No': 0, 'No internet service': 0, 'No phone service':0,
'Month-to-month': 2 , 'One year': 1, 'Two year':0, # тем выше значение, чем больше внимания уделяется оплате 
 'Credit card (automatic)': 0,                     # требует меньше всего участия для проведения оплаты
 'Bank transfer (automatic)': 1,                   # банки уведомляют об автоплаже заранее, увеличивают участие клиента 
  'Mailed check': 2,                               # требует больше всего усилий, но это клиенты, оплачивающие по старинке
  'Electronic check': 3,                           # требует много усилий и внимания со стороны клиента
  'DSL': 1, 'Fiber optic': 2,                      # DSL лучше, чем ничего, но хуже оптики
  'Male': 1, 'Female': 0}

data.drop(['Unnamed: 0', 'customerID'], axis=1, inplace=True)
data = data.replace(replace_dict).replace(' ', 0).astype('float', errors='ignore')
data.head()
```

# 2. Исследование зависимостей и формулирование гипотез

Мы видим множество категориальных значений, которые подробно описывают объем услуг, которые использует клиент. За исключением демографических признаков, мы можем условно разделить фичи на три группы, которые вероятно связаны с фактом оттока схожим образом.

Гипотеза 1. Чем больше усилий клиенту приходится прикладывать для проведения оплаты, тем чаще он задумывается о соответствии качества услуг их цене. Вовлеченность в процесс оплаты будем считать по способу оплаты, типу контракта, безбумажному билингу 

Гипотеза 2. Существует определённый период, по достижению которого увеличивается риск потерять клиента.

Гипотеза 3. Сервисы связанные с безопасностью характеризуют клиента как человека осторожного, это хорошо если клиент уже сотрудничает с нами. Потребность в цифровой безопасности включает в себя онлайн безопасность, бэкап, страховку, техподдержку. 

Демографические характеристики, телефонию и время сотрудничества  пока не будем определять в какие-то группы. При необходимости, сделаем это позже.


```python
# выведем попарное сравнение числовых признаков
sns.pairplot(data[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']], 
                 hue='Churn',  size=4)
```

```python
# давайте посмотрим на корреляцию фичей с главной характеристикой
data.astype(float).corr()['Churn'].sort_values()
```

Итак мы уже можем сделать некоторые промежуточные выводы:    
Больше всего с уходом клиента связан тип контракта, тип интернет-соединения, метод оплаты и время сотрудничества. Чем активнее участие клиента в оплате услуг и чем лучше тип предоставляемого соединения и меньше срок сотрудничества, тем больше вероятность этого клиента потерять.
Если рассматривать в целом, то мы видим следующее:
1. Пользование услугами, связанными с потребностью с безопасностью имеют отрицательную корреляцию с шансом потерять клиента.
2. Вовлеченность в процесс оплаты имеет положительную корреляцию.
3. Из потребности в развлекательных услугах наибольшее значение при оценке рисков имеет тип интернет соединения.
4. Какого-то определенного периода, после которого начинает расти риск ухода - нет. Но самый первый месяц связан с наибольшими потерями клиентов.

Кроме того, мы видим, что отрицательную корреляцию также имеют характеристики: партнер и иждивенцы. Пол, и телефония не оказывают значительного влияния, их мы рассматривать не будем.

Мы знаем на какие характеристики стоит обратить больше всего внимания, поэтому давайте посмотрим их связь с другими фичами

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(14,12))
plt.title('Матрица корреляций параметров клиентского профиля', y=1.05, size=15)
sns.heatmap(data.astype(float).corr(),linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)
plt.show()
```

Мы видим, что больше всего интернет связан со стриминговыми сервисами и стоимостью услуг. Однако эта корреляция обязательно должна была проявить себя т.к. стриминг сервисы могут быть только при наличии интернета.  Характеристики вовлеченности в оплату связаны между собой и имеют отрицательную корреляцию с временем сотрудничества, партнером и иждивенцами. 

Хотя мы не обнаружили значимую связь времени сотрудничества и демографических характеристик с потребностью в безопасности, мы можем определить, что все эти характеристики "не рискового" клиента и нам надо более детально рассмотреть потенциального "беглеца".

```python
# посмотрим в какие промежутки времени мы чаще всего теряем 
# в отличии от sns.pairplot посмотрим эти значения на дискретных барах
# сгруппируем по месяцам и посчитаем количество потерь, а потом выведем топ 10 месяцев
risk_period = data.groupby('tenure', as_index=False)\
    .agg({'Churn':'sum'}).sort_values('Churn', ascending=False)\
    .reset_index(drop=True)['tenure'][:10].to_list()

# построим график для этого периода
sns.countplot(data[data['tenure'].isin(risk_period)==True]['tenure'], hue=data['Churn'])
plt.show()
```

```python
# выведем графики для остальных наиболее значемых характеристик 
for i in ['PaymentMethod', 'Contract', 'PaperlessBilling', 'InternetService']:
    sns.countplot(data[i], hue=data['Churn']  )
    plt.show()
```

Сейчас мы можем определить потрет клиента, которого легко потерять:     
1. Это новый клиент, самый опасный месяц - первый.
2. Клиент получает счета каждый месяц в электронном формате и оплачивает счета самостоятельно, также через интернет.
3. Клиент подключил соединение через оптоволокно.

Мы не знаем точные причины почему именно такой клиент вероятнее всего уйдёт.Но мы можем сформировать несколько гипотез, исходя из того портрета, который имеем,  для дальнейшего изучения:    
Гипотеза 1. Клиенты, которые приходят в компанию за высокоскоростным интернетом, разочарованы качеством предоставляемых услуг.

Гипотеза 2. При попытке провести оплату услуг через сайт компании, возникает ошибка, которая либо не позволяет завершить оплату вообще, либо отталкивает новых клиентов от дальнейшего сотрудничества.

Гипотеза 3. Клиенты попробовали услуги по промо-тарифу и не захотели проводить продлевать услуги, когда узнали размер последующих платежей.

# 3. Построение моделей для прогнозирования оттока

В данном случае перед нами стоит задача бинарной классификации. Мы будем использовать 3 классификатора чтобы собрать с них прогнозы и обучить на них метамодель. Так как зависимости, которые мы определили были выявлены с помощью метода df.corr() из коробки, то это значит, что мы нашли связь с помощью линейного коэффициента корреляции Пирсона. Следовательно, самый очевидный метод обучения для нашего случая - это классификатор линейной регрессии. Также обучим модель градиентного бустинга и метод опорных векторов. Таким образом для метамодели у нас будет набор данных полученных от трех разных классификаторов.

Датасет не сбалансирован, но выборка не слишком большая, чтобы мы могли просто выбрасывать лишние значения, поэтому метрика должна быть устойчива к неравному соотношению классов.
Метрика ROC AUC используется для измерения производительности для задач классификации при различных настройках пороговых значений и обладает нужной нам устойчивостью.

```python
# вызовем библиотеки
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import recall_score, accuracy_score, roc_auc_score
from xgboost import XGBClassifier

# укажем подбор гиперпараметров  для наших моделей
gbc_params = {'learning_rate': np.arange(0.1, 0.6, 0.1)}  

svc_params = {'kernel': ['rbf'], # SVC
              'C': np.arange(0.01, 1, 0.03)}

lr_params = {'C': np.arange(0.05, 1, 0.05)}

xgbc_params = {
            'max_depth': range (2, 10, 1),
            'n_estimators': range(60, 220, 40),
            'learning_rate': [0.1, 0.01, 0.05]
                }
# запакуем в список все параметры 
params = [gbc_params, svc_params, lr_params, xgbc_params]
gbc, svc, logreg, meta = GradientBoostingClassifier(), SVC(),  LogisticRegression(), XGBClassifier()
models = [gbc, svc, logreg]
data_train, data_test, targets_train,  targets_test = train_test_split(data.drop(['Churn'], axis=1),
                                                    data.Churn.values, test_size=0.3, random_state=17, stratify=data.Churn.values)

x_train, x_test, y_train, y_test = train_test_split(data_train, targets_train, test_size=0.3, random_state=17, stratify=targets_train)  
meta_mtrx = np.empty((x_test.shape[0], len(models)))
meta_mtrx_test = np.empty((data_test.shape[0], len(models)))
for n, model in enumerate(models):
    grid_model = GridSearchCV(model, params[n], refit=True, cv=5, scoring='roc_auc')
    grid_model = grid_model.fit(x_train, y_train)
    meta_mtrx[:, n] = grid_model.predict(x_test);
    meta_mtrx_test[:, n] = grid_model.predict(data_test)
meta_model = GridSearchCV(meta, params[n+1], cv=5, scoring='roc_auc').fit(meta_mtrx, y_test)
meta_predict = meta_model.predict(meta_mtrx_test)
print('roc_auc_score: ',roc_auc_score(targets_test, meta_predict), 
      '\nrecall_score:', recall_score(targets_test, meta_predict), 
      '\naccuracy_score: ', accuracy_score(targets_test, meta_predict))
# метапрогноз получен, можем перейти к сравнению качества
```

# 4. Сравнение качества моделей 

Метрика площади под кривой будет использоваться для понимания целостного результата модели. Количество пойманых уходов клиента, при настройке порогового значения, будем определять метрикой recall. Также выведем значения для базовых моделей и на  основании их результатов применим настройки порогового значения.

```python
# выставим значения для прогноза, где при вероятности ухода выше 0.2 засчитываем уход
new_pred = meta_model.predict_proba(meta_mtrx_test)[:,0]
new_pred = [0 if x > 0.8 else 1 for x in new_pred]
print('roc_auc_score: ',roc_auc_score(targets_test, new_pred), 
      '\nrecall_score:', recall_score(targets_test, new_pred), )
```

```python
#сделаем тоже самое, но с порогом 0.1
new_pred = meta_model.predict_proba(meta_mtrx_test)[:,0]
new_pred = [0 if x > 0.9 else 1 for x in new_pred]
print('roc_auc_score: ',roc_auc_score(targets_test, new_pred), 
      '\nrecall_score:', recall_score(targets_test, new_pred), )
```

Как мы видим модель имеет четкие пороговые границы и такая калибровка неэффективна. Применим туже логику на базовых моделях

```python
gbc, svc, logreg, meta = GradientBoostingClassifier(), SVC(),  LogisticRegression(), XGBClassifier()
```

```python
gbc_model = GridSearchCV(gbc, gbc_params, cv=5, scoring='roc_auc').fit(data_train, targets_train)
gbc_pred = gbc_model.predict(data_test)
logreg_model = GridSearchCV(logreg, lr_params, cv=5, scoring='roc_auc').fit(data_train, targets_train)
logreg_predict = logreg_model.predict(data_test)
svc_model = GridSearchCV(svc, svc_params, cv=5, scoring='roc_auc').fit(data_train, targets_train)
svc_predict = svc_model.predict(data_test)
xgbc_model = GridSearchCV(meta, xgbc_params, cv=5, scoring='roc_auc').fit(data_train, targets_train)
xgbc_pred = xgbc_model.predict(data_test)

print('градиентныей бустинг xgboost- roc_auc:', roc_auc_score(targets_test, xgbc_pred),
      'recall:',recall_score(targets_test, xgbc_pred),)
print('градиентныей бустинг sklearn - roc_auc:', roc_auc_score(targets_test, gbc_pred),
      'recall:', recall_score(targets_test, gbc_pred),)
print('линейная регрессия - roc_auc:', roc_auc_score(targets_test, logreg_predict),
      'recall:', recall_score(targets_test, logreg_predict), )
print('метод опорных векторов - roc_auc:', roc_auc_score(targets_test, svc_predict),
      'recall:', recall_score(targets_test, svc_predict), )
```

Здесь мы видим, что модели обученные через GridSearch показывают лучшее значение чем метамодель. Кроме модели опорных векторов, все её предсказания приняли значение 0. Для остальных изменим пороговое значение и выведем результат:

```python
# изменим порог значений с 0.5 до 0.2 для модели градиентного бустинга sklearn
gbc_new = gbc_model.predict_proba(data_test)[:,1]
gbc_new = np.array([1 if x > 0.15 else 0 for x in gbc_new])
print('градиентный бустинг sklearn- recall:',
      recall_score(targets_test, gbc_new),
      'roc_auc_score:', roc_auc_score(targets_test, gbc_new),)
```

```python
# изменим порог значений с 0.5 до 0.2 для модели логистической регресси
logreg_new = logreg_model.predict_proba(data_test)[:,1]
logreg_new = np.array([1 if x > 0.15 else 0 for x in logreg_new])
print('логистическая регрессия - recall:',
      recall_score(targets_test, logreg_new),
      'roc_auc_score:', roc_auc_score(targets_test, logreg_new),)
```
